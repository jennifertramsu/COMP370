```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Defining variables
path = "../data"
file = path + "/nyc_311_limit_2020_filtered.csv"
```

```python
!python ../src/borough_complaints.py -i $file -s 01/01/2020 -e 02/28/2020 -o $path/complaint_types_01_02.csv
```

```python
# Determine most abundant complaint type
df = pd.read_csv(path + "/complaint_types_01_02.csv")
df_sum = df.groupby('complaint type')['count'].sum().sort_values(ascending=False)
abundant = df_sum.index[0]

print("The most abundant complaint type is: " + abundant)
```

```python
df_abundant = df[df['complaint type'] == abundant]
plt.bar(df_abundant['borough'], df_abundant['count'])
plt.title(abundant + ' Complaints Across Boroughs from Jan-Feb')
plt.xlabel('Borough')
plt.ylabel('Count')
```
